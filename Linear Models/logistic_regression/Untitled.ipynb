```python
x = data.drop(["Survived"], axis = 1)
y = data["Survived"]
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
train_x, test_x, train_y, test_y = tts(x,y ,random_state = 56)
```

```python
#importing Logistic Regression and metric F1-score
from sklearn.linear_model import LogisticRegression as LogReg
from sklearn.metrics import f1_score
```

```python
# Creating instance of Logistic Regresssion
logreg = LogReg()

# # Fitting the model
logreg.fit(train_x, train_y)
```

```python
train_predict =  logreg.predict(train_x)
```

```python
train_predict
```

```python
tain_score = f1_score(train_predict, train_y)
```

```python
tain_score
```

```python
prob = logreg.predict_proba(train_x)
```

```python
prob
```

```python
train_probapility = prob[:,1]
```

```python
for i in range (0, len(train_probapility)):
    if (train_probapility[i] > 0.5):
        train_probapility[i] = 1
    else: 
        train_probapility[i] = 0
```

```python
train_probapility
```

```python
# calculating f1 score
train_f1_score = f1_score(train_probapility, train_y)

print("the train F1_score is " ,train_f1_score)
```

```python

test_predict =   logreg.predict(test_x)
test_f1_score = f1_score(test_predict, test_y)
print("the test f1 Score is ",test_f1_score )
```
