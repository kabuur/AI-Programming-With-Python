```python
data = pd.read_csv("./big_mart_sales.csv")
```

```python
data.isna().sum()
```

```python
from sklearn.impute import SimpleImputer as spi
```

```python
imput_weight = spi(strategy="mean")
imputer_size = spi(strategy= "mean")
```

```python
imput_weight.fit(data[["Item_Weight"]])
data.Item_Weight = imput_weight.transform(data[["Item_Weight"]])
```

```python
imputer_size.fit(data[["Item_Weight"]])
data.Outlet_Size = imputer_size.transform(data[["Item_Weight"]])
```
