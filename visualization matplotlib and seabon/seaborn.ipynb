```python
import seaborn as sb
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./bigmart_data.csv")
```

```python
data
```

```python

sb.set()
sb.set(style="darkgrid")
plt.rcParams["figure.figsize"] = (10,10)
```

```python
data.dropna(how="any", inplace=True)
```

```python
# data.isna().sum()
```

```python
data["Visibility_scaled"] = data["Item_Visibility"] *100
```

```python
data
```

# Line plot

```python
sb.lineplot(data=data[:50])
```

```python
sb.barplot(x="Item_Type", y= "Item_MRP", data = data[:50])
```

```python
sb.distplot(data["Item_MRP"])
```
